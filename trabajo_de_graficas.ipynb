```python
# Grafica 1 Ventas

from matplotlib import legend
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
import numpy as np


datos1  = {
    "Producto": ["A", "B", "C", "D"],
    "Ventas": [50, 30, 70, 20]
}

# Convertimos el diccionario en un DataFrame
df1 = pd.DataFrame(datos1)

# Crear una nueva figura de grafica
plt.figure()

# Creamos la gráfica de barras y cambiamos el color
plt.bar(df1["Producto"], df1["Ventas"], color="skyblue")

# Agregamos título y etiquetas
plt.title("Ventas por Producto")
plt.xlabel("Producto")
plt.ylabel("Ventas")

# Mostramos la gráfica
plt.show()
```

```python
# Grafica 2 Ganancias

from matplotlib import legend
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
import numpy as np

datos  = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril"],
    "Ganancias": [100, 150, 130, 170]
}

df2 = pd.DataFrame(datos)

plt.figure()

# Creamos gráfica de línea y cambiamos color
plt.plot(df2["Mes"], df2["Ganancias"], color="green", marker="o")

plt.title("Ganancias Mensuales")
plt.xlabel("Mes")
plt.ylabel("Ganancias")

plt.show()
```

```python
# Grafica 2 Temperaturas

from matplotlib import legend
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
import numpy as np

datos = {
    "Ciudad": ["Bogota", "Medellin", "Cali", "Cartagena"],
    "Temperatura": [12, 18, 20, 24]
}
df3 = pd.DataFrame(datos)

plt.figure()

# Creamos gráfica circular con colores personalizados
plt.pie(
    df3["Temperatura"],
    labels=df3["Ciudad"],
    colors=["red", "blue", "orange", "purple"],
    autopct="%1.1f%%"
)

plt.title("Temperaturas por cada ciudad")

plt.show()
```
